```python
def get_the_urls_list(url):
    suburl = []
    subtitle = []
    response = requests.get(url)
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')
    
        # Find all divs with class "content"
        contents = soup.find_all('div', class_='content')
    
        # Loop through each content div and extract details
        for idx, content in enumerate(contents, start=1):
            # Find all <a> elements with the class "HgbW6-m" inside the content
            links = content.find_all('a', class_='HgbW6-m')
    
            if links:  # Only process content that contains at least one link
                # Ignore the first link and get the second link (if it exists)
                if len(links) > 1:
                    second_link = links[1]
                    suburl.append("https://hellobacsi.com" + second_link.get('href'))
                    subtitle.append(second_link.text)
    else:
        print(f"Failed to fetch the page. Status code: {response.status_code}")
    return suburl, subtitle
```

```python
def split_into_chunks(text, max_seq_len, tokenizer):
    tokens = tokenizer.encode(text, add_special_tokens=False)  # Tokenize without special tokens
    chunks = [tokens[i:i + max_seq_len] for i in range(0, len(tokens), max_seq_len)]  # Split into chunks
    decoded_chunks = [tokenizer.decode(chunk) for chunk in chunks]  # Decode tokens back to text
    return decoded_chunks

# Apply the function to the DataFrame
df["chunks"] = df["content"].apply(lambda x: split_into_chunks(x, 256, tokenizer))
```

```python
import numpy as np

def retrieve_relevant_chunks(question, qdrant_client, collection_name, model, top_n):
    # Encode the question to a vector
    question_vector = model.encode(question, show_progress_bar=False).tolist()
    # Perform the search
    search_results = qdrant_client.query_points(
        collection_name=collection_name,
        query=question_vector,
        limit=top_n,
        with_payload=True,
        score_threshold=0.5
    )
    # Extract and return the 'content' field from the payload
    contents = [result.payload['content'] for result in search_results.points]
    return contents
    
def retriever(input_text):
    # Replace with your collection name and model
    collection_name = "doctor_v2"
    retrieved_chunks = retrieve_relevant_chunks(input_text, qdrant_client, collection_name, model, 4)
    return "\n\n".join([chunk for chunk in retrieved_chunks])
```

```python
def extract_section_data(url):
    """
    Fetch the HTML from a given URL and extract section titles and content.

    Args:
        url (str): The URL of the webpage.

    Returns:
        list[dict]: A list of dictionaries, each containing 'title' and 'content' of a section.
    """
    try:
        # Fetch the HTML content
        response = requests.get(url)

        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')
        # Find all sections
        sections = soup.find_all('section')
        # Extract titles and content
        results = []
        for section in sections:
            # Extract title
            title_element = section.find('h2', class_='TopicFHead_topicFHead__upuEK professional sticky-fhead readable')
            title = title_element.get_text(strip=True) if title_element else "No Title"

            # Extract content
            content_element = section.find('div', class_='TopicFHead_fHeadBody__mLB6h')
            content = content_element.get_text(strip=True) if content_element else "No Content"

            # Append to results
            if title != "No Title" and content != "No Content" and title != "Thông tin thêm":
                results.append({'title': title, 'content': content})

        return results

    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL: {e}")
        return []
    except Exception as e:
        print(f"Error processing HTML: {e}")
        return []
```

```python
trimmer = trim_messages(
    max_tokens = 10,
    strategy="last",
    token_counter=len,
    include_system=True,
    allow_partial=True,
    start_on="human",
)

class State(TypedDict):
    messages: Annotated[Sequence[BaseMessage], add_messages]
    
def call_model(state: State):
    # Trim the messages using the trimmer
    trimmed_messages = trimmer.invoke(state["messages"])
    # Construct the prompt with the trimmed messages
    formatted_prompt = prompt.invoke(
        {"context": retriever(user_query),  # Pass the context to the template
        "input": user_query,   # Replace with the actual user query
        "messages": trimmed_messages,
        "patient_info": patient_info}  # Include the trimmed messages
    )
    # Call the LLM with the formatted prompt
    response = llm.invoke(formatted_prompt)
    # Append the AI response to the state
    return {"messages": [response]}

workflow = StateGraph(state_schema=State)
workflow.add_edge(START, "model")
workflow.add_node("model", call_model)

memory = MemorySaver()
app = workflow.compile(checkpointer=memory)
```

```python
import underthesea

def chunk_content_vietnamese(content, model, max_seq_len = 256):
    sentences = underthesea.sent_tokenize(content)
    chunks = []
    current_chunk = []
    current_chunk_len = 0
    
    for sentence in sentences:
        sentence_len = len(model.tokenize([sentence])["input_ids"][0])
        if current_chunk_len + sentence_len > max_seq_len:
            chunks.append(" ".join(current_chunk))
            current_chunk = []
            current_chunk_len = 0
        current_chunk.append(sentence)
        current_chunk_len += sentence_len
    
    if current_chunk:
        chunks.append(" ".join(current_chunk))
    
    return chunks
```

```python
def encode_text(text, model):
    return model.encode(text, show_progress_bar=False).tolist()

def process_doctor_collection(qdrant_client, model, max_seq_len, target_collection):
    points = unique_all_points[13163:]
    index = 13159
    for point in points:
        content = point.payload["chunks"]
        title = point.payload["title"]
        url = point.payload["url"]
        if not title or not content:
            continue
        # Tokenize the content
        content_len = len(model.tokenize([title + ' ' + content])["input_ids"][0])
        
        if content_len <= max_seq_len:
            # Content fits in one chunk
            vector = encode_text(title + ' ' + content, model)
            new_point = {
                "id": index,
                "vector": vector,
                "payload": {"title": title, "content": title + ' ' + content, "url": url}
            }
            qdrant_client.upsert(collection_name=target_collection, points=[new_point])
            index += 1
        else:
            # Content needs to be chunked
            chunks = chunk_content_vietnamese(title + ' ' + content, model, max_seq_len)
            for i, chunk in enumerate(chunks):
                vector = encode_text(chunk, model)
                new_point = {
                    "id": index,
                    "vector": vector,
                    "payload": {"title": title, "content": title + ' ' + chunk, "url": url}
                }
                qdrant_client.upsert(collection_name=target_collection, points=[new_point])
                index += 1
```
